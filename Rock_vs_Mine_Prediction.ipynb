WORK FLOW----> Data collection --> Data Preprocessing --> Model Training --> Output

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data= pd.read_csv('/content/Copy of sonar data.csv',header=None)
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data[60].value_counts()
```

```python
x=sonar_data.drop(columns=60,axis=1)
y=sonar_data[60]
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

**Model Evalution**

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_data_score=accuracy_score(x_train_prediction,y_train)
print(training_data_score*100)
```

```python
x_test_prediction=model.predict(x_test)
testing_data_score=accuracy_score(x_test_prediction,y_test)
print(testing_data_score*100)
```
